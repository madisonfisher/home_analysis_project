```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
sales_df = pd.read_csv("median_sale_price.csv")
sales_df.head()
```

```python
list_df = pd.read_csv("median_list_price.csv")
list_df.head()
```

```python
sales_df_us = sales_df[sales_df['RegionName'] == "United States"]
sales_df_us.head()
```

```python
sales_df_us = sales_df_us.drop(columns=["SizeRank","RegionID","RegionType","StateName"])
sales_df_us.head()
```

```python
list_df_us = list_df[list_df['RegionName'] == "United States"]
list_df_us.head()
```

```python
list_df_us = list_df_us.drop(columns=["SizeRank","RegionID","RegionType","StateName"])
list_df_us.head()
```

```python

list_df_us.set_index('RegionName',inplace=True)
list_df_us = list_df_us.transpose()

list_df_us.head()
```

```python

sales_df_us.set_index('RegionName',inplace=True)
sales_df_us = sales_df_us.transpose()
sales_df_us.head()
```

```python
list_df_us.index = pd.to_datetime(list_df_us.index)
list_df_us = list_df_us.rename(columns={"United States":"United_States"})
list_df_us.head()
```

```python
sales_df_us = sales_df_us[(sales_df_us.index > '2018-01-01')]
sales_df_us = sales_df_us.rename(columns={"United States":"United_States"})
sales_df_us.head()

```

```python
sales_df_us.index = pd.to_datetime(sales_df_us.index)
sales_df_us.index
```

```python

data_6m_rol = sales_df_us["United_States"].rolling(window = 6, center = True).mean().to_frame()

data_6m_rol
```

```python
fig, ax = plt.subplots(figsize = (11,4))
# plotting monthly data
ax.plot(sales_df_us['United_States'], marker='.', markersize=12, color='0.6',linestyle='None', label='Monthly')
# plotting 3-month rolling mean
ax.plot(data_6m_rol["United_States"], linewidth=2, label='6-M Rolling Mean')
# Beautification of plot

ax.legend()
ax.set_xlabel('Year')
ax.set_ylabel('Price(USD)')
ax.set_title('Trends in Home Sale Prices')
```

```python
list_6m_rol = list_df_us["United_States"].rolling(window = 6, center = True).mean().to_frame()

list_6m_rol
```

```python
fig, ax = plt.subplots(figsize = (11,4))
# plotting monthly data
line = ax.plot(list_df_us['United_States'], marker='.', markersize=12, color='0.6',linestyle='None', label='Monthly Data')
# plotting 3-month rolling mean
ax.plot(list_6m_rol["United_States"], linewidth=2, label='6-M Rolling Mean(List)')
# Cleaning up plot
ax.plot(sales_df_us['United_States'], marker='.', markersize=12, color='0.6',linestyle='None')
# plotting 3-month rolling mean
ax.plot(data_6m_rol["United_States"], linewidth=2, label='6-M Rolling Mean(Sale)')

plt.grid()

ax.legend()
ax.set_xlabel('Year-Month')
ax.set_ylabel('Price(USD)')
ax.set_title('Median Home List/Sale Prices')


fig.savefig("Images/SalevsList.pdf")
```
