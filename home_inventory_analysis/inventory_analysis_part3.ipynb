```python
places_count = len(mean_locations)
for j in range(0,places_count):
    mean_location_df = mean_pending.iloc[:, [1,(2+j)]]
    mean_location_df = mean_location_df.rename(columns={f"{mean_locations[j]}": "Mean_Pending"})
    mean_location_df[["Mean_Pending"]] = mean_location_df[["Mean_Pending"]].apply(pd.to_numeric)
    mean_location_df.loc[:,'LastWeek'] = mean_location_df.loc[:,"Mean_Pending"].shift()
    mean_location_df.loc[:,'LastWeek_Diff'] = mean_location_df.loc[:,'LastWeek'].diff()
    mean_location_df.loc[:,'2Weeks'] = mean_location_df.loc[:,"LastWeek"].shift()
    mean_location_df.loc[:,'2Weeks_Diff'] = mean_location_df.loc[:,'2Weeks'].diff()
    mean_location_df = mean_location_df.dropna()
    mean_location_df = mean_location_df.reset_index(drop=True)
    X = mean_location_df[:].drop(['Mean_Pending'], axis = 1)
    y = mean_location_df["Mean_Pending"]
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=57)
    from sklearn.linear_model import LinearRegression
    model = LinearRegression()
    model.fit(X_train, y_train)
    r = len(mean_location_df.index)
    r = r - 1
    #running half a year of predicted data
    for i in range(0,25):
        LastWeekValue = mean_location_df["Mean_Pending"][r]
        TotalDaysValue = mean_location_df["Total_Days"][r] + 7
        LastWeekDifferenceValue = mean_location_df["Mean_Pending"][r] - mean_location_df["LastWeek"][r]
        TwoWeekValue = mean_location_df["LastWeek"][r]
        TwoWeekDifferenceValue = mean_location_df["LastWeek"][r] - mean_location_df["2Weeks"][r]
        NewX = [[TotalDaysValue, LastWeekValue, LastWeekDifferenceValue, TwoWeekValue, TwoWeekDifferenceValue]]
        predicted_value = model.predict(NewX)
        predicted_number = str(predicted_value).lstrip('[').rstrip(']')
        predicted_pending = float(predicted_number)
        new_row = {'Total_Days': TotalDaysValue, 'Mean_Pending': predicted_pending, 'LastWeek': LastWeekValue, 'LastWeek_Diff': LastWeekDifferenceValue, "2Weeks" :TwoWeekValue, "2Weeks_Diff" : TwoWeekDifferenceValue}
        mean_location_df = mean_location_df.append(new_row, ignore_index = True)
        r = r + 1
        i = i + 1
    mean_location_final = mean_location_df[["Total_Days", "Mean_Pending"]]
    mean_location_final = mean_location_final.rename(columns={"Mean_Pending": mean_locations[j]})
    if j == 0:
        final_mean_pending = mean_location_final
    else:
        final_mean_pending = pd.merge(final_mean_pending, mean_location_final, on=["Total_Days", "Total_Days"], how="outer")
    j = j + 1
final_mean_pending = final_mean_pending.fillna('') 
        
```
